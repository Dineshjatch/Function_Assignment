Que 1. What is the difference between a function and a method in Python?
  - A function is an independent block of code that performs a specific task and is defined using def. It can be called without being tied to an object.
  - Example:
        def greet(name):
          return f"Hello, {name}!"

        print(greet("Dinesh"))  

  - A method is a function that is associated with an object and is called using the object's instance. Methods are defined inside a class and operate on the class's attributes.
  - Example:
          class Greeter:
            def greet(self, name):
              return f"Hello, {name}!"

          greeter = Greeter()
          print(greeter.greet("Dinesh"))

Que 6. Explain the concept of generators in Python and how they are defined.
  - A generator is a special type of iterator that allows you to iterate over data without storing it in memory. It generates values on the fly using the yield keyword instead of return.
  - Generators are defined like normal functions but use yield instead of return.
  - Example:
        def my_generator():
          yield 1
          yield 2
          yield 3

        gen = my_generator()
        print(next(gen))  
        print(next(gen))
        print(next(gen))  

Que 7.  What are the advantages of using generators over regular functions?
  - - Memory Efficient = Generates values one at a time instead of storing them all in memory.

    - Lazy Evaluation = Values are computed only when needed, avoiding unnecessary processing.

    - Better Performance = Faster execution, especially for large datasets.

    - Automatic Iterators = No need for __iter__() and __next__() methods.

    - Supports Pipelines = Can be chained for efficient data processing.
    - Example:
          def gen_numbers():
            for i in range(3):
                yield i

          def list_numbers():
              return [i for i in range(3)]

          print(list(gen_numbers()))  
          print(list_numbers())  

Que 8. What is a lambda function in Python and when is it typically used?

  - Example:

  - We use lambda function for inline function like map(), filter() and sorted(). For temporary use.

Que 9. Explain the purpose and usage of the `map()` function in Python.
  - - Applies a function to all elements in an iterable.

    - Returns an iterator, saving memory.

    - Useful for transformations like type conversion or mathematical operations.
    - Example:
          numbers = [1, 2, 3, 4]
          squares = list(map(lambda x: x * x, numbers))
          print(squares)  

Que 10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
  - The map() function applies a given function to each element of an iterable and returns an iterator with transformed values. It is used for data transformation.

  - The filter() function selects elements from an iterable based on a condition and returns an iterator with only the elements that satisfy the condition. It is used for data filtering.

  - The reduce() function applies a function cumulatively to all elements of an iterable, reducing it to a single value. It is used for aggregation and requires the functools module.

  - Example:
        numbers = (1, 2, 3, 4)

        squares = map(lambda x: x * x, numbers)
        print(tuple(squares))  

        evens = filter(lambda x: x % 2 == 0, numbers)
        print(tuple(evens))  

        from functools import reduce
        sum_numbers = reduce(lambda x, y: x + y, numbers)
        print(sum_numbers)

```python
#4. Write a Python function that checks if a given number is prime or not from 1 to 200.
def check_prime(n):
  for i in range(2, 200):
    if n%i==0:
      return False
    else:
      return True
print(check_prime(int(input("Enter a number between 1 to 200: "))))
```

```python
#5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.
# In this module and in previous modules, there is no lecture of class.
def fib(n):
  a=0
  b=1
  for i in range(n):
    yield a
    a,b=b,a+b

f=fib(10)
print(f)
print(next(f))
print(next(f))
print(next(f))
print(next(f))
```

```python
#6. Write a generator function in Python that yields the powers of 2 up to a given exponent.
def power_of_2(n):
  for i in range(n):
    yield 2**i
p=power_of_2(10)
print(p)
print(next(p))
print(next(p))
print(next(p))
print(next(p))
```

```python
#10. Create a Python program that uses `filter()` to remove all the vowels from a given string.
vowels=['a', 'i', 'e', 'o', 'u', 'A', 'E', 'I', 'O', 'U']
string=input("Enter String: ")
filtered_string=filter(lambda x: x not in vowels, string)
print(list(filtered_string))
```
